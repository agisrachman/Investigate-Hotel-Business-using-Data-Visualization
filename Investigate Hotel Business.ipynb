## Copy Dataframe

```python
df_clean = df.copy() #copy dataframe for Cleaning
```

## Missing Value

```python
#detect missing values
df_clean.isnull().sum()
```

There are 4 columns with missing data : children, city, agent, company.

### Checking missing values

```python
#cheking missing values for children
df_clean.loc[df_clean['children'].isnull(), : ]
```

```python
#Cheking missing values for city
df_clean.loc[df_clean['city'].isnull(), : ]
```

```python
#cheking missing values for agent and company
df_clean[df_clean['agent'].notnull() & df_clean['company'].notnull()]
```

### Handling missing values

```python
#for childreen : deleted
df_clean.dropna(subset=['children'],inplace=True)
df_clean['children'].isnull().sum() #check the total missing value
```

since the number of missing values is small, the row is deleted.

```python
#for city : deleted
df_clean.dropna(subset=['city'],inplace=True)
df_clean['city'].isnull().sum() #check the total missing value
```

since the number of missing values is small, the row is deleted.

```python
#for agent and company : delete column
df_clean.drop(columns=['company','agent'], axis=1, inplace=True)
```

In addition to many missing values, the column is not included for analysis.

## Duplicate data

### Checking duplicate data

```python
#Total duplicate data
df_clean.duplicated().sum()
```

```python
#Display duplicate data
df_clean.loc[df_clean.duplicated(),:]
```

### Handling duplicate data

```python
df_clean.drop_duplicates(keep='last', inplace=True) #Remove duplicate data, by retrieving the last data
df_clean.duplicated().sum() ##Total duplicate data
```

## Data suitability

```python
#calculate the sum of values
for col in df_clean:
    print(f'''Value count column {col}:''')
    print(df_clean[col].value_counts())
    print('___'*20, '\n')
    print()
```

```python
print(df_clean['meal'].value_counts()) #Calculating unique values
print(print('___'*20, '\n'))
print(df_clean['distribution_channel'].value_counts()) #Calculating unique values
```

There are undefined values (meal, distribution_channel)

```python
df_clean['meal'] = df_clean['meal'].replace('Undefined' , ) #Replace according to the previous row
df_clean['distribution_channel'] = df_clean['distribution_channel'].replace('Undefined' , ) #Replace according to the previous row
print(df_clean['meal'].value_counts()) #Calculating unique values
print(print('___'*20, '\n'))
print(df_clean['distribution_channel'].value_counts()) #Calculating unique values
```

```python
df_clean.drop(columns=['arrival_date_week_number', 'required_car_parking_spaces'], axis=1, inplace=True) #remove unrelated columns
```

delete columns that are not needed

```python
print("before pre-processing: ", df.shape) #view the number of rows and columns before cleaning
print("after pre-processing: ", df_clean.shape) #view the number of rows and columns after cleaning
```
